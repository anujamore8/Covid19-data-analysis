__Import Libraries__

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

__Load Clean Dataset__

```python
df = pd.read_csv("covid_cleaned_dataset.csv")
df.head()
```

__Basic Data Understanding__

__Check structure of dataset.__

```python
df.shape
df.info()
df.describe()
```

```python
df.isnull().sum()
```

__Global COVID Trends Over Time__

__This shows how cases increased worldwide.__
__Identify when the pandemic peaked.__


```python
global_cases = df.groupby("date")["new_cases"].sum()

plt.figure(figsize=(12,6))
plt.plot(global_cases)
plt.title("Global COVID Daily Cases Trend")
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.show()
```

__Top 10 Most Affected Countries__

__Identify the most affected countries.__

```python
top_cases = df.groupby("location")["total_cases"].max().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_cases.values, y=top_cases.index)
plt.title("Top 10 Countries with Highest COVID Cases")
plt.xlabel("Total Cases")
plt.ylabel("Country")
plt.show()
```

__Country-wise Death Analysis__

```python
top_deaths = df.groupby("location")["total_deaths"].max().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_deaths.values, y=top_deaths.index)
plt.title("Top 10 Countries with Highest Deaths")
plt.xlabel("Total Deaths")
plt.ylabel("Country")
plt.show()
```

__Death Rate Comparison__

__Create new column.__

```python
df["death_rate"] = (df["total_deaths"] / df["total_cases"]) * 100
```

__Top countries with highest death rate:__

```python
death_rate = df.groupby("location")["death_rate"].max().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=death_rate.values, y=death_rate.index)
plt.title("Countries with Highest Death Rate")
plt.xlabel("Death Rate (%)")
plt.show()
```

__Vaccination Trends__

__Global vaccination progress.__

__Shows how vaccination increased over time.__

```python
vaccination_trend = df.groupby("date")["people_vaccinated"].sum()

plt.figure(figsize=(12,6))
plt.plot(vaccination_trend)
plt.title("Global Vaccination Trend")
plt.xlabel("Date")
plt.ylabel("People Vaccinated")
plt.show()
```

__Cases vs Deaths Relationship__

__Helps understand correlation between cases and deaths.__

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="total_cases", y="total_deaths", data=df)
plt.title("Cases vs Deaths Relationship")
plt.show()
```

__Correlation Heatmap__

__Shows relationships between variables.__

```python
cols = [
    "total_cases",
    "new_cases",
    "total_deaths",
    "new_deaths",
    "people_vaccinated",
    "total_vaccinations"
]

plt.figure(figsize=(8,6))
sns.heatmap(df[cols].corr(), annot=True, cmap="coolwarm")
plt.title("Important Feature Correlation")
plt.show()
```

__Continent-wise Case Comparison__



```python
continent_cases = df.groupby("continent")["total_cases"].max()

continent_cases.plot(kind="bar", figsize=(8,5))
plt.title("COVID Cases by Continent")
plt.xlabel("Continent")
plt.ylabel("Total Cases")
plt.show()
```

__Boxplot (Outlier Detection)__

__Insight:__

__Detect extreme values (e.g., USA, India)__

__Shows distribution skewness__

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df[["total_cases","total_deaths"]])
plt.title("Outlier Detection")
plt.show()
```

__Violin Plot (Distribution Shape__

__Insight:__

__Shows density + distribution__

__Better than histogram__

```python
plt.figure(figsize=(10,6))
sns.violinplot(data=df["new_cases"])
plt.title("Distribution of Daily Cases")
plt.show()
```

__Top Countries Trend Line__

__Compare country-wise growth patterns__

```python
top5 = df.groupby("location")["total_cases"].max().sort_values(ascending=False).head(5).index

top_df = df[df["location"].isin(top5)]

plt.figure(figsize=(12,6))
sns.lineplot(data=top_df, x="date", y="total_cases", hue="location")
plt.title("Top 5 Countries COVID Growth")
plt.show()
```

__Vaccination vs Deaths__

__Shows impact of vaccination__

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="people_vaccinated", y="total_deaths", data=df)
plt.title("Vaccination vs Deaths")
plt.show()
```
